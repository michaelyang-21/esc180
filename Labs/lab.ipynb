```python
text = ["I", "am", "a", "sick", "man.", "I", "am", "a", "spiteful", "man.", "I", "am",
        "an", "unattractive", "man.", "I", "believe", "my", "liver", "is", "diseased.",
        "However,", "I", "know", "nothing", "at", "all", "about", "my", "disease,", "and",
        "do", "not", "know", "for", "certain", "what", "ails", "me."]



def count_words(text):
    word_counts = {}
    for word in text:
        if word == '':
            continue
        if word not in word_counts:
            word_counts[word] = 1
        else:
            word_counts[word] += 1
    return word_counts

print(count_words(text))
print(count_words(text)['man.'])
```

```python
from random import randrange


def top10(L):
    return sorted(L)[::-1][0:10]


def random_list(n, a, b):
    L = []
    for i in range(n):
        L.append(randrange(a, b))
    return L


L = random_list(100, 0, 101)

print(L)
print(top10(L))
```

```python
def top10_words(counts):
    L = []

    for item in counts.items():
        L.append((item[1], item[0]))

    return sorted(L)[::-1][0:10]

print(top10_words(count_words(text)))
```

```python
p_and_p = open("1342-0.txt", encoding="latin-1").read()

punctuation = ['.', ',', ':', ';', '"', '(', ')', '\n', '_']

for punc in punctuation:
    p_and_p = p_and_p.replace(punc, ' ')

p_and_p = p_and_p.split(' ')

print(top10_words(count_words(p_and_p)))
```

```python
import urllib.request


def find_search_results(url):
    f = urllib.request.urlopen(url)
    page = f.read().decode("utf-8")
    f.close()

    end = page.find('search results')
    start = end

    while page[start] != '>':
        start -= 1

    str = page[start+1:end+14]
    str = str.split(' ')

    return int(str[1].replace(',', ''))


among_results = find_search_results(
    'https://ca.search.yahoo.com/search?p=among+us')

print(among_results)
```

```python
def yahooify(search):
    return f'https://ca.search.yahoo.com/search?p={search}'


def choose_variant(variants):
    results = []
    for variant in variants:
        num = (find_search_results((yahooify(urllib.parse.quote(variant)))), variant)

        results.append(num)

    return sorted(results)[::-1][0]


print(choose_variant(['"five-year anniversary"', '"fifth anniversary"']))

print(choose_variant(['top ranked school uoft', 'top ranked school waterloo']))
```

```python
results = open("results.txt", encoding='utf8').read()

results = results.split('\n')

school_count = {}

for result in results:
    a = result.split(',')
    a = a[0].split(' ')
    if not a[0].isupper():
        continue
    b = ''
    for i in a:
        if i.isupper():
            continue
        else:
            b += i + ' '
    b = b[0:-1]

    if b not in school_count:
        school_count[b] = 1
    else:
        school_count[b] += 1

def sort_dict(dict):
    L = []

    for item in dict.items():
        L.append((item[1], item[0]))

    return sorted(L)[::-1]
        

for i in sort_dict(school_count):
    print(f'{i[1]: <60} | {i[0]}')
```
